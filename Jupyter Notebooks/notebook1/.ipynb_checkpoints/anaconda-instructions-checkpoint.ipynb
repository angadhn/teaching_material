
## Python and Jupyter Notebook

In this course we will being learning some basic computer programming in the programming language *Python*.  In particular we will be using *Jupyter Notebook*.

The cleanest way to get started is to install the free *Anaconda* distribution of Python (see instructions below).  This will give you all that you need for this course, including Jupyter Notebook.

If you are unable to install Anaconda, you could consider Google Colab (which requires Google account) or Jupyter Lab (see instructions below).



### Installing Anaconda

Go to [Anaconda Individual Edition](https://www.anaconda.com/products/individual).  This is free and open-source softare so will not cost you anything.

Click on the Download button which take you to the bottom of the webpage and a list of Anaconda Installers for different operating systems (Windows, MacOS or Linux).



* For Windows, once everything is installed, you can go to the Windows Start Menu and look in the Anaconda3 menu and choose Jupyter Notebook (Anaconda3).  This will bring up a console window that will start the system running, but the Jupyter Notebook will eventually appear in your **web browser**.

* For MacOS, _insert instructions here_.

Once you have Jupyter Notebook up and running,
click on the "New" button/menu (far right-hand side) and select "Python 3".  This will give a new blank Jupyter Notebook.

![Jupyter Notebook blank](jupyter01.png)

In the box next "In [ ]:] type
```{python}
print("Hello, World!")
```
Then click on the Run button in the menu bar (below Cell) to run the Python code in that box.  Below is what we should expect to see.

![Jupyter Notebook Hello World](jupyter02.png)

[Hello, World!](https://en.wikipedia.org/wiki/%22Hello,_World!%22_program) is a tradition in Computer Science as the first thing a student learns when learning a new programming language.

Now you are all set up with Jupyter Notebook on your local machine.




### Google Colab

If you have a Google account, a simple way to get up and running on Jupyter Notebook without installing anything is to use the Google Colaboratory (called Google Colab for short).

Make sure you are using one of Chrome, Firefox or Safari web browsers (it doesn't seem to work with Microsoft Edge).

Go to [Google Colab](https://colab.research.google.com/).  Note that you will need a Google account to log in and store your files on Google Drive.

Click on "NEW NOTEBOOK" at the bottom of the popup window or go to the File Menu and select "New notebook".

In the box next to the arrow, type
```{python}
print("Hello, World!")
```
then click on the arrow to the left of the box to run the Python code in that box.  Below is what we should expect to see.

![Colab Hello World](colab01.png)

Now you can continue to use Jupyter Notebook hosted on Google Colab.
